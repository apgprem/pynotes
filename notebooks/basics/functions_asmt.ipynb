```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

print(factorial(5))  # Output: 120
```

    for i in range(6, n+1):
        result *= i
        return result

```python
def count_vowels(s: str):
    return sum(1 for char in s if char.lower() in "aeiou")
print(count_vowels("Hello World"))
```
