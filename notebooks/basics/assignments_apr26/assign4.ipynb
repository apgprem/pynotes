pythonevens_only = [n for n in numbers if n % 2 == 0]

```python
pythonnumbers = [12, 7, 25, 3, 18, 9, 30, 14, 6, 21]

for i in pythonnumbers:
    if i%2 == 0:
        print("Even Number", i)
    else:
        print("Odd Number", i)
    
```
