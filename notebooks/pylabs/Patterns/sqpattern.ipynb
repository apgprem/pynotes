```python
def print_hollow_square(n):
# n=int(input("enter no"))

    for i in range(n):
        for j in range(n):
            if i==0 or i==n-1 or j==0 or j==n-1:
                print("*",end=" ")
            else:
                print(" ",end=" ")
        print()
print_hollow_square(5)
```

```python
# Hollow Trianngle - Pattern

n=int(input("enter No"))

for i in range(n):
    for j in range(i+1):
        if j==0 or i==n-1 or i==j:
            print("* ",end=" ")
        else:
            print("  ",end=" ")
    print(" ")
```

```python
# Right Trianngle - Pattern

n=int(input("enter No"))

for i in range(n,0,-1):
    for j in range(i):
        print("*",end=" ")
    print(" ")
```

```python
# Right Trianngle - Pattern

n=int(input("enter No"))

for i in range(n-2):
    for j in range(n):
        print("*",end=" ")
    print(" ")
    
```
